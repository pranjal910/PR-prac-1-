```python
import pandas as pd
```

```python
data = {
    'pH': [6.5, 8.5, 6.6, 6.6, 6.5, 6.6, 6.5, 6.8, 6.6, 6.7],
    'Temprature': [36, 70, 43, 50, 38, 45, 37, 45, 45, 45],
    'Taste': [0, 0, 1, 0, 1, 0, 0, 0, 0, 1],
    'Odor': [0, 0, 0, 0, 0, 0, 0, 1, 0, 1],
    'Turbidity': [0, 0, 1, 1, 0, 1, 0, 1, 1, 0],
    'Colour': [247, 246, 250, 250, 255, 250, 255, 255, 250, 245],
    'Grade': ['medium', 'low', 'high', 'low', 'medium', 'medium', 'medium',
       'high', 'medium', 'medium']
    
}
```

```python
df = pd.DataFrame(data)
```

MAPPING LOW MEDIUM AND HIGH AS 0 1 2 RESPECTIVELY

```python
df
```

```python
df['Grade']=df['Grade'].map({'low':0,'medium':1,'high':3})
```

```python
df
```

```python
X = df.drop(['Grade'],axis=1)
y = df['Grade']
```

```python
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
svm = SVC(gamma = 'auto')
gnb = GaussianNB()
```

```python
n = int(input('Enter an option\n 0. Exit\n 1. SVM\n 2. Naive Bayes\n '))


class Classifier():
    def svm():
        svm.fit(X, y)
        print('The accuracy of SVM is ',svm.score(X, y))
        print('The predicted values are ', list(svm.predict(X)))
        print('\n')

    def gnb():
        gnb.fit(X, y)
        print('The accuracy of Gaussian Naive Bayes is ',gnb.score(X, y))
        print('The predicted values are ', list(gnb.predict(X)))
        print('\n')



while True:
    if n == 0:
        break
    elif n == 1:
        Classifier.svm()
    elif n == 2:
        Classifier.gnb()
    n = int(input('Enter an option\n 0. Exit\n 1. SVM\n 2. Naive Bayes\n '))
```
